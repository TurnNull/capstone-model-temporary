```python
import os
os.environ['TF_USE_LEGACY_KERAS'] = '1'
import pandas as pd
import tensorflow as tf
import tensorflowjs as tfjs
```

```python
# 1. Baca file parquet Anda
df_train = pd.read_parquet('grand_final_predictions_train.parquet')
df_val = pd.read_parquet('grand_final_predictions_val.parquet')
```

```python
# 2. Pisahkan Fitur (Kolom hasil prediksi 5 model) dan Target (Nilai Asli/Aktual)
X_train = df_train.drop(columns=['target_column_name'])
y_train = df_train['target_column_name']

X_val = df_val.drop(columns=['target_column_name'])
y_val = df_val['target_column_name']
```

```python
# 3. Bangun model TensorFlow Linear
meta_model_tf = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(X_train.shape[1],), activation=None)
])
```

```python
# 4. Compile dan Latih Model
meta_model_tf.compile(optimizer='adam', loss='mse', metrics=['mae'])
meta_model_tf.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=32)
```

```python
# 5. Simpan langsung ke format TensorFlow.js!
tfjs.converters.save_keras_model(meta_model_tf, './tfjs_meta_learner')
print("Model Grand Final Meta-Learner berhasil disimpan untuk TF.js!")
```
