```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
BASE_PATH = 'dataset'
METRICS_FOLDER = os.path.join(BASE_PATH, 'results')
```

```python
lr_metrics = pd.read_csv(
    os.path.join(METRICS_FOLDER, 'linear_regression_metrics.csv')
)

ridge_metrics = pd.read_csv(
    os.path.join(METRICS_FOLDER, 'ridge_regression_metrics.csv')
)

rf_metrics = pd.read_csv(
    os.path.join(METRICS_FOLDER, 'random_forest_metrics.csv')
)

xgb_metrics = pd.read_csv(
    os.path.join(METRICS_FOLDER, 'xgboost_metrics.csv')
)
```

```python
# gabungkan semua metrics
baseline_metrics = pd.concat([
    lr_metrics,
    ridge_metrics,
    rf_metrics,
    xgb_metrics
], ignore_index=True)

print("BASELINE MODEL COMPARISON")
print(baseline_metrics)
```

```python
# Visualisasi compare R²
plt.figure(figsize=(8,5))

plt.bar(
    baseline_metrics['model'],
    baseline_metrics['test_r2']
)

plt.title('Perbandingan Test R² Model')
plt.ylabel('R² Score')
plt.xticks(rotation=10)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# simpan hasil akhir
baseline_metrics.to_csv(
    os.path.join(METRICS_FOLDER, 'baseline_metrics.csv'),
    index=False
)

print("\nbaseline_metrics.csv berhasil disimpan.")
```
